```r
install.packages("BiocManager")
BiocManager::install("seqinr")
```

```r
library("seqinr")
choosebank() # List all sub-databases in ACNUC
```

```r
choosebank("genbank")
query("SchistosomamRNA", "SP=Schistosoma mansoni AND M=mRNA")
closebank()
```

```r
choosebank("refseqViruses")
Dengue1 <- query("dengue1", "AC=NC_001477") #list variable Dengue1
```

```r
attributes(Dengue1)
```

```r
Dengue1$nelem # $ to retrieve value of named element
Dengue1$req
```

```r
dengueseq <- getSequence(Dengue1$req[[1]])
dengueseq[1:50]
```

```r
annots <- getAnnot(Dengue1$req[[1]])
annots[1:20]
closebank()
```

```r
choosebank("genbank")
HumtRNAs <- query("humtRNAs", "SP=homo sapiens AND M=TRNA")
myseqs <- getSequence(HumtRNAs)
mynames <- getName(HumtRNAs)
```

```r
write.fasta(myseqs, mynames, file.out="humantRNAs.fasta")
closebank()
```

```r
attributes(Rabiesvirus)
Rabiesvirus$req
```
